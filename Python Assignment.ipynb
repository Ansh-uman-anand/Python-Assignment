```python
#3 Compare and contrast mutable and immutable objects in Python with examples.

#Ans. 
Mutable:
Mutable objects can be changed after creation, meaning their contents can be modified. Lists, dictionaries, and sets are prime examples.

Immutable:
Immutable objects cannot be changed after creation. Integers, strings, and tuples are examples of immutable objects.

Examples of Mutable Objects:

list = [1, 2, 3]
list[0] = 10  # Modifies the first element of the list
print(list)  # Output: [10, 2, 3]

Examples of Immutable Objects:

x = 10
y = x
x = 20  # x now points to a new integer object
print(y)  # Output: 10 (y still points to the original object)
```

```python
#4  Discuss the different types of operators in Python and provide examples of how they are used.

Ans. Python divides the operators in the following groups:

1 Arithmetic operators -    +, - , *, /, %, **, //
2 Assignment operators -    =, +=, -=, *=, /=, %=, **=, //=
3 Comparison operators -    ==, !=, >, < , >=, <=
4 Logical operators -       and , or , not
5 Identity operators -      is, is not
6 Membership operators -    in, in not
7 Bitwise operators -       &, ^, ~, <<, >>

Arithmetic operators example

a = 10
b = 3
print(a + b)  # Output: 13
print(a - b)  # Output: 7
print(a * b)  # Output: 30
print(a / b)  # Output: 3.3333333333333335
print(a % b)  # Output: 1
print(a ** b) # Output: 1000
print(a // b) # Output: 3

Assignment operators example

a = 5
a += 3  # a = a + 3
print(a)  # Output: 8
a *= 2  # a = a * 2
print(a)  # Output: 16


Comparison operators example

a = 10
b = 3
print(a == b)  # Output: False
print(a != b)  # Output: True
print(a > b)   # Output: True
print(a < b)   # Output: False
print(a >= b)  # Output: True
print(a <= b)  # Output: False


Logical operators example

a = True
b = False
print(a and b)  # Output: False
print(a or b)   # Output: True
print(not a)    # Output: False


Identity operators example

a = [1, 2, 3]
b = a
c = [1, 2, 3]
print(a is b)      # Output: True (b references the same object as a)
print(a is c)      # Output: False (c is a different object with the same content)
print(a is not c)  # Output: True


Membership operators example

a = [1, 2, 3, 4]
print(2 in a)       # Output: True
print(5 not in a)   # Output: True


Bitwise operators example

a = 5  # 0101 in binary
b = 3  # 0011 in binary
print(a & b)  # Output: 1 (0001 in binary)
print(a | b)  # Output: 7 (0111 in binary)
print(a ^ b)  # Output: 6 (0110 in binary)
print(~a)     # Output: -6 (Two's complement)
print(a << 1) # Output: 10 (Shifts left by 1 bit)
print(a >> 1) # Output: 2 (Shifts right by 1 bit)
```

```python
#5 Explain the concept of type casting in Python with examples.

Ans. 
   int(): Converts a value to an integer.
   float(): Converts a value to a float.
   str(): Converts a value to a string.
   list(): Converts a value (such as a tuple) to a list.
   tuple(): Converts a value (such as a list) to a tuple.
   
   Converting String to Integer:

   num_str = "100"
   num_int = int(num_str)  # Explicitly converting string to integer
   print(num_int)  # Output: 100
   print(type(num_int))  # Output: <class 'int'>
   
   Converting Float to Integer:

   x = 9.8
   y = int(x)  # Explicitly converting float to integer
   print(y)  # Output: 9
   
   
  Converting Integer to String:
  
  x = 50
  y = str(x)  # Explicitly converting integer to string
  print(y)  # Output: '50'
  print(type(y))  # Output: <class 'str'>
 
 
  Converting List to Tuple:

  my_list = [1, 2, 3]
  my_tuple = tuple(my_list)  # Explicitly converting list to tuple
  print(my_tuple)  # Output: (1, 2, 3)
```

```python
#6 How do conditional statements work in Python? Illustrate with examples.
 
 Ans. Conditional statements in Python are used to make decisions based on certain conditions. They allow you to      execute specific blocks of code depending on whether a condition is True or False. The primary conditional          statements in Python are if, elif, and else.
 
   'if' Statement:
   The if statement checks a condition and executes the block of code if the condition is True.

   Syntax:
          if condition:
          # code to execute if condition is True
   
   Example: 
           age = 18
           if age >= 18:
           print("You are eligible to vote.")

    'elif' Statement:
                   The elif (short for "else if") statement allows you to check multiple conditions if the previous      conditions are False. You can have multiple elif statements following an if.
    
   Syntax:
          if condition1:
          # code to execute if condition1 is True
          elif condition2:
          # code to execute if condition2 is True
          
  Example: 
          score = 85
          if score >= 90:
                print("Grade: A")
          elif score >= 80:
                print("Grade: B")
          elif score >= 70:
                print("Grade: C")
          else:
                print("Grade: D")
                
   'else' Statement:
                  The else statement is executed when all preceding conditions are False. It is optional and should          be the last block in a conditional statement.            
               
   Syntax:
           if condition:
           # code to execute if condition is True
           else:
           # code to execute if condition is False
   Example: 
   
           temperature = 15
           if temperature > 20:
               print("It's warm outside.")
           else:
               print("It's cold outside.")
```

```python
#7 Describe the different types of loops in Python and their use cases with examples.

Ans:
    In Python, loops are used to repeatedly execute a block of code as long as a specified condition is met. Python provides several types of loops to handle different scenarios. The main types of loops in Python are for loops and while loops.
    
    for Loop:
     The for loop is used to iterate over a sequence (like a list, tuple, string, or range) and execute a block of      code for each item in the sequence.

     Syntax:
     for variable in sequence:
    # code to execute for each item
   
    Example:
    fruits = ["apple", "banana", "mango"]
    for fruit in fruits:
    print(fruit)
    
    Output
    apple
    banana
    mango
    
    
       # Using range() to Generate Numbers:-
       
      for i in range(5):
      print(i)
      Output
      0
      1
      2
      3
      4

       # Iterating Over Characters in a String:

      word = "Ajay"
      for letter in word:
      print(letter)
      Output
      A
      j
      a
      y
      
 while Loop:
        The while loop executes a block of code as long as a specified condition is True. It is useful when the          number of iterations is not known beforehand.
        
   Syntax: 
   
    while condition:
    # code to execute while condition is True
    
    Exapmle:
    
    count = 0
    while count < 5:
        print(count)
        count += 1

    Output:
    0
    1
    2
    3
    4
    
 
 
 # Nested Loops:
   We can nest loops inside one another to perform complex operations. Both for and while loops can be nested.
   
   Syntax:
   
   for variable1 in sequence1:
       for variable2 in sequence2:
           # code to execute

  Example:
      for i in range(3):
          for j in range(2):
              print(f"i = {i}, j = {j}")
  
  Output:
   i = 0, j = 0
   i = 0, j = 1
   i = 1, j = 0
   i = 1, j = 1
   i = 2, j = 0
   i = 2, j = 1
```
